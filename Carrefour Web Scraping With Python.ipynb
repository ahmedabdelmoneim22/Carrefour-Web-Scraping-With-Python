```python
# my-Libraries.
# Scrape information from Web Pages.
from bs4 import BeautifulSoup
# Interacting with Web APIs or Web scraping.
import requests
# SMTP Protocol Client Send Emails.
import smtplib
import time
import datetime
```

```python
# Connect to Carrefour Website.
URL = 'https://www.carrefouregypt.com/mafegy/en/smartphones/vivo-y17s-6gb-128gb-forest-green/p/624571?list_name=search%7CiPhone&offer=offer_carrefour_'
# Define user-agent to mimic a real browser.
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9'
}
getRequest = requests.get(URL,headers=headers)
# reading The Page in HTML Using BeautifulSoup().
Soup1 = BeautifulSoup(getRequest.content,"html.parser")
print("Ensure it Work",getRequest.status_code)
print(Soup1)
```

```python
#Search Google:-Vivo Y17S - 6.56-inches 128GB/6GB Dual SIM 4G Smartphone - Forest Green.
#>>prettify():-Prettify the parsed content.    
#&Get Specific Elements.
Soup2 =  BeautifulSoup(Soup1.prettify(),"html.parser")
print(Soup2)
```

```python
#>>Go TO T-Shirt Carrefour Page Choose Title Right-Click>>Inspect elements.
#<h1 class="css-106scfp">Vivo Y17S - 6.56-inches 128GB/
#6GB Dual SIM 4G Smartphone - Forest Green</h1>
Title = Soup2.find(class_="css-106scfp").get_text().replace(" ", "")
print(Title)#Print-The-Text.
```

```python
# Choose Location Egypt.
# Click on the Price>>right-Click>>Inspect elements.
selectPrice = Soup2.find(class_="css-17ctnp").get_text()
Price=selectPrice.split()[1]
```

```python
import datetime
today = datetime.date.today()
print(today)
print(Title)
print(Price)
```

```python
import csv
header = ['Title','Price','Date']
data = [Title,Price,today]
with open('CarrefourWebScrapingWithPython.csv','w',newline='',encoding='UTF8') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerow(data)
```

```python
def Get_Price():
    URL = 'https://www.carrefouregypt.com/mafegy/en/smartphones/vivo-y17s-6gb-128gb-forest-green/p/624571?list_name=search%7CiPhone&offer=offer_carrefour_'
    # Define user-agent to mimic a real browser.
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
               'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9'}
    getRequest = requests.get(URL,headers=headers)
    # reading The Page in HTML Using BeautifulSoup().
    Soup1 = BeautifulSoup(getRequest.content,"html.parser")
    Soup2 =  BeautifulSoup(Soup1.prettify(),"html.parser")
    Title = Soup2.find(class_="css-106scfp").get_text().replace(" ", "")
    selectPrice = Soup2.find(class_="css-17ctnp").get_text()
    Price=selectPrice.split()[1]
    import datetime
    today = datetime.date.today()
    import csv
    header = ['Title','Price','Date']
    data = [Title,Price,today]
    with open('CarrefourWebScrapingWithPython.csv','+a',newline='',encoding='UTF8') as file:
        writer = csv.writer(file)
        writer.writerow(data)
```

```python
# Every Day Get The Price of Item;
while(True):
    Get_Price()
    time.sleep(86400)#60x60x24=Day;
```
